```python
import json
with open('.MNIST_Results/loss_data_RMSProp.json', 'r') as f:
    loaded_loss_RMSProp = json.load(f)

with open('.MNIST_Results/loss_data_Adams.json', 'r') as f:
    loaded_loss_Adams = json.load(f)

with open('.MNIST_Results/loss_data_SGDm.json', 'r') as f:
    loaded_loss_SGDm = json.load(f)

with open('.MNIST_Results/loss_data_FishLeg.json', 'r') as f:
    loaded_loss_FishLeg = json.load(f)
```

```python
import json
import matplotlib.pyplot as plt

# Plot the data
plt.figure(figsize=(10, 5))

# Plot training loss
plt.subplot(1, 2, 1)
plt.plot(loaded_loss_RMSProp['training_loss_RMSProp'], label='RMSProp')
plt.plot(loaded_loss_Adams['training_loss_ADAMS'], label='Adams')
plt.plot(loaded_loss_SGDm['training_loss_SGDm'], label='SGDm')
plt.plot(loaded_loss_FishLeg['training_loss_FishLeg'], label='FishLeg')
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plot testing loss
plt.subplot(1, 2, 2)
plt.plot(loaded_loss_RMSProp['testing_loss_RMSProp'], label='RMSProp')
plt.plot(loaded_loss_Adams['testing_loss_ADAMS'], label='Adams')
plt.plot(loaded_loss_SGDm['testing_loss_SGDm'], label='SGDm')
plt.plot(loaded_loss_FishLeg['testing_loss_FishLeg'], label='FishLeg')
plt.title('Testing Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```
